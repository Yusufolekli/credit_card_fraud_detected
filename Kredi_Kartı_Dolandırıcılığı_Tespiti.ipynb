```python
#Analiz için verilerin ayrıştırılması
legit = credit_card_data[credit_card_data.Class == 0]
fraud=  credit_card_data[credit_card_data.Class == 1]
```

Under-Sampling


---


Normal işlemlerin ve dolancırıcılık işlemlerinin benzer dağılımını içeren örnek bir veri kümesi oluşturmak.

dolandırıcılık işlemi sayısı -> *492*

```python
legit_sample = legit.sample(n=492)
```

iki dataframenin birleştirmesi

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

verileri özelliklere ve hedeflere bölme

```python
X = new_dataset.drop(columns= 'Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Verileri eğitim verileri ve test verileri olarak ikiye ayırın.

---


amaç = modelin aşırı öğrenmeden kaçınarak (overfitting) genelleme yapabilme yeteneği ölçülür.

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Eğitimi

Logistik Regresyon

```python
model = LogisticRegression(solver='liblinear', max_iter=200)
```

```python
#loji̇sti̇k regresyon modeli̇ni̇n eği̇ti̇m veri̇leri̇yle eği̇ti̇lmesi̇
model.fit(X_train, Y_train)
```

```python
#eğitim verilerinde doğruluk (accuracy)
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
```

```python
print('Eğitim verilerinde ki Accuracy: ', training_data_accuracy )
```

```python
#test verilerinde ki accuracy
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Test verilerinde ki Accuracy: ', test_data_accuracy)
```
